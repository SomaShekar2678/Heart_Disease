```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x = data.drop('target', axis=1)
y = data['target']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train, y_train)

y_pred_lr = lr.predict(x_test)

cf_lr = confusion_matrix(y_test, y_pred_lr)
acc_lr = accuracy_score(y_test, y_pred_lr)
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
y_pred_dt = dt.predict(x_test)

cf_dt = confusion_matrix(y_test, y_pred_dt)
ac_dt = accuracy_score(y_test, y_pred_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier()
rf.fit(x_train, y_train)
y_pred_rf = rf.predict(x_test)

cf_rf = confusion_matrix(y_test, y_pred_rf)
ac_rf = accuracy_score(y_test, y_pred_rf)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
y_pred_kn  = knn.predict(x_test)
cf_kn = confusion_matrix(y_test, y_pred_kn)
ac_kn = accuracy_score(y_test, y_pred_kn)
```

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
y_pred_xg = xgb.predict(x_test)
cf_xg = confusion_matrix(y_test, y_pred_xg)
ac_xg = accuracy_score(y_test, y_pred_xg)
```

```python

print("Logistic Regression Confusion Matrix:\n", cf_lr)
print("Decision Tree Confusion Matrix:\n", cf_dt)
print("Random Forest Confusion Matrix:\n", cf_rf)
print("K-Nearest Neighbors Confusion Matrix:\n", cf_kn)
print("XGBoost Confusion Matrix:\n", cf_xg)
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

def plot_confusion_matrix(cf_matrix, model_name):
  plt.figure(figsize=(6, 4))
  sns.heatmap(cf_matrix, annot=True, fmt='g', cmap='Blues')
  plt.title(f'Confusion Matrix for {model_name}')
  plt.xlabel('Predicted Label')
  plt.ylabel('True Label')
  plt.show()


plot_confusion_matrix(cf_lr, "Logistic Regression")
plot_confusion_matrix(cf_dt, "Decision Tree")
plot_confusion_matrix(cf_rf, "Random Forest")
plot_confusion_matrix(cf_kn, "K-Nearest Neighbors")
plot_confusion_matrix(cf_xg, "XGBoost")
```

```python


print("Logistic Regression Classification Report:\n", classification_report(y_test, y_pred_lr))
print("Decision Tree Classification Report:\n", classification_report(y_test, y_pred_dt))
print("Random Forest Classification Report:\n", classification_report(y_test, y_pred_rf))
print("K-Nearest Neighbors Classification Report:\n", classification_report(y_test, y_pred_kn))
print("XGBoost Classification Report:\n", classification_report(y_test, y_pred_xg))
```

```python


from sklearn.metrics import recall_score, precision_score, f1_score

# Logistic Regression
recall_lr = recall_score(y_test, y_pred_lr)
precision_lr = precision_score(y_test, y_pred_lr)
f1_lr = f1_score(y_test, y_pred_lr)

# Decision Tree
recall_dt = recall_score(y_test, y_pred_dt)
precision_dt = precision_score(y_test, y_pred_dt)
f1_dt = f1_score(y_test, y_pred_dt)

# Random Forest
recall_rf = recall_score(y_test, y_pred_rf)
precision_rf = precision_score(y_test, y_pred_rf)
f1_rf = f1_score(y_test, y_pred_rf)

# K-Nearest Neighbors
recall_kn = recall_score(y_test, y_pred_kn)
precision_kn = precision_score(y_test, y_pred_kn)
f1_kn = f1_score(y_test, y_pred_kn)

# XGBoost
recall_xg = recall_score(y_test, y_pred_xg)
precision_xg = precision_score(y_test, y_pred_xg)
f1_xg = f1_score(y_test, y_pred_xg)


print("Logistic Regression - Recall:", recall_lr, "Precision:", precision_lr, "F1-score:", f1_lr)
print("Decision Tree - Recall:", recall_dt, "Precision:", precision_dt, "F1-score:", f1_dt)
print("Random Forest - Recall:", recall_rf, "Precision:", precision_rf, "F1-score:", f1_rf)
print("K-Nearest Neighbors - Recall:", recall_kn, "Precision:", precision_kn, "F1-score:", f1_kn)
print("XGBoost - Recall:", recall_xg, "Precision:", precision_xg, "F1-score:", f1_xg)
```

```python

model_accuracies = {
    "Logistic Regression": acc_lr,
    "Decision Tree": ac_dt,
    "Random Forest": ac_rf,
    "K-Nearest Neighbors": ac_kn,
    "XGBoost": ac_xg
}

best_model = max(model_accuracies, key=model_accuracies.get)
best_accuracy = model_accuracies[best_model]

print(f"The best model is {best_model} with an accuracy of {best_accuracy:.4f}")

def select_top_n_models(model_accuracies, n=2):
    sorted_models = sorted(model_accuracies.items(), key=lambda item: item[1], reverse=True)
    top_n_models = [model_name for model_name, accuracy in sorted_models[:n]]
    return top_n_models

top_models = select_top_n_models(model_accuracies, n=2)
print(f"The top models are: {top_models}")
```

```python


plt.figure(figsize=(5, 5))
plt.bar(model_accuracies.keys(), model_accuracies.values())
plt.title('Model Accuracies')
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
